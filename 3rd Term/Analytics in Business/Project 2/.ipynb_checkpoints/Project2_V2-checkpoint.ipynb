```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
def import_data(csv_name):
    csv_pd = pd.read_csv(csv_name)
    csv_pd.Timestamp = pd.to_datetime(csv_pd.Timestamp)
    csv_pd['Year'] = csv_pd.Timestamp.dt.year
    csv_pd['Month'] = csv_pd.Timestamp.dt.month
    csv_pd['Day'] = csv_pd.Timestamp.dt.day
    csv_pd['Hour'] = csv_pd.Timestamp.dt.hour
    csv_pd['Minute'] = csv_pd.Timestamp.dt.minute
    csv_pd['Second'] = csv_pd.Timestamp.dt.second
    return csv_pd
residential_demand = import_data('residential_cleaned.csv')
turbine_telemetry = import_data('turbine_telemetry.csv')
```

# Data cleaning:
### The following parts is to add additional columns to turbine_telemetry for future data analysis

```python
# Remove the rows with nan values. 

turbine_telemetry_replace_nan = turbine_telemetry.copy()
Power_Wind_pd = turbine_telemetry_replace_nan[['Timestamp','Power_kw','Wind_ms']]
Setpoint_pd = turbine_telemetry_replace_nan[['Timestamp','Setpoint_kw']]
Power_Wind_pd.fillna(method='bfill',inplace=True)
Setpoint_pd.fillna(method='ffill',inplace=True)
turbine_telemetry_replace_nan = pd.merge(turbine_telemetry_replace_nan, Power_Wind_pd, on=["Timestamp"])
turbine_telemetry_replace_nan = pd.merge(turbine_telemetry_replace_nan, Setpoint_pd, on=["Timestamp"])
turbine_telemetry_replace_nan=turbine_telemetry_replace_nan.drop(columns = ['Power_kw_x','Setpoint_kw_x','Wind_ms_x'])
turbine_telemetry_replace_nan['Timestamp_minus_sc'] = turbine_telemetry_replace_nan['Timestamp'].dt.floor('T')
turbine_telemetry_replace_nan = turbine_telemetry_replace_nan.drop_duplicates(subset=['Timestamp_minus_sc'], keep='last')
turbine_telemetry_replace_nan=turbine_telemetry_replace_nan.rename(columns={'Power_kw_y':'Power_kw','Wind_ms_y':'Wind_ms','Setpoint_kw_y':'Setpoint_kw'})
turbine_telemetry_replace_nan = turbine_telemetry_replace_nan.drop(columns = ['Timestamp_minus_sc'])
turbine_telemetry_replace_nan.head()
```

```python
# Add Half-hourly timestamp column in turbine telemetry to match the timestamp in residential_demand
import datetime
turbine_telemetry_replace_nan['Timestamp_H'] = turbine_telemetry_replace_nan['Timestamp'].apply(lambda dt: datetime.datetime(dt.year, dt.month, dt.day, dt.hour,30*(dt.minute // 30)))
turbine_telemetry_replace_nan.head()
```

Model:

n is the number of household who join in the DR program(based on data)

p_energy is the energy used for heating that was original supplied by gas but now supplied by the curtailed energy  (per household) (based on data)

P1 is the price that we buy the curtailed energy from the generators (wholesale price, based on online data)

P2 is the price that we sell the curtailed energy to the local, which are used for heating (estimated)

P3 is the price of gas used to supply the heater (based on online data)

FC is the installment cost of this DR program (to be determined)

VC is the variable cost per household of this PR program (to be determined)


Revenue for generator: n * p_energy  * P1 

Revenue for each household:  p_energy  * (P3 - P2) 

Financial benefits to Kaluza: Revenue - cost =  n * p_energy  * (P2-P1) -  FC - VC * n

Savings to the Community: when the demand and supply are not balanced, it will hurt the grid. Therefore, we need to balance the usage of electronicty during peak and offpeak time and when wind is strong and not strong.


Hence for the next step, given the DR system is 15.4KWh and 2.2 KW => which require charging hour for 7hours
    1. We are ging to assume the hot water system is going to be charged in mid of the day -> increase electricity demand and reduce curtailment
    2.  DR will support residential electricity usage at night -> reduce electricity demand and increase curtailment
 
HOWEVER: 
    1. If at night & without DR system involved, energy is already oversupplied (curtailment>0) -> replacing residential electricity demand by DR system will lead to new curtailment energy!
    2. This means partial of our saved curtailed energy from mid of the day will be wasted at night!
    3. Hence the overall curtailment energy saving only comes from time periods at night that there is no electricity curtailment (curtailment == 0)
    

Hence energy saving from DR system is calculated as follow:
    For % of household that have DR system installed:
            1. During time period that DR system is supporting residential electricity usage (at night)
            2. For each half hour that energy is curtailed, DR system will not support residential electricity usage.
            3. For each half hour that energy is not curtailed, DR system is going to support residential electricity usage.
            4. As such, energy curtailment saving = Sum (residential electricity demand, if energy is not curtailed during this period)
