## Exploratory Data Analysis

### Create lookup table
Let's see how we can do some data cleaning and deal with outliers in our dataset.

```python
# how to figure out what's the the smallest speed necessary to produce power
# there's a lot of similar thinking between Pandas and SQL
# .iloc (by row number) vs .loc (by index value, or column)
v_min = tt[(tt.loc[:, 'Power_kw'] >= 10.)].loc[:, 'Wind_ms'].quantile(0.1) # 10% quantile
v_min
```

```python
# Some examples:
# we only want to use data when the turbine was not constrained
# there are two options here -- queries and filter 
# query is more SQL-ish
df = tt.query('Setpoint_kw==900').copy()

# remove data points when the wind speed is fast enough for power production but there is no power; turbine is probably manually switched off here
df.loc[((df.Wind_ms >= v_min) & (df.Power_kw.abs() <= 1e-5))] = np.nan

# remove all missing values from the dataframe
df = df.dropna(0, inplace=False)
```

```python
# let's attempt the scatter plot again
df.plot.scatter('Wind_ms', 'Power_kw', alpha=0.01, figsize=(15, 5)) # a lot cleaner!
```

```python
# select bin locations
# from the data we can see that after about 16m/s, the power output is nearly at max
# NOTE: you could achieve a similar doing a quantiles query
wmax_ = df[df.Power_kw == df.Power_kw.quantile(0.90)].Wind_ms.quantile(0.5)
print(wmax_)
wmax = 17
```

```python
bins = arange(0, wmax, .2)

# group data by bins
# "advanced" Pandas feature
cuts = pd.cut(df.Wind_ms, bins) # <-- generating bins for the windspeep
gp = df.Power_kw.groupby(cuts) # iterator

# hack -- for visualisation purposes only
print()
for n, g in gp:
    print(n, g)
    break
```

```python
# "elementary" solution in case you're interested 
bins_ = [round(_, 2) for _ in arange(0, wmax, .2)] 
bins = list(zip(bins_, bins_[1:])) # each one of them corresponds to a "group"
bins[:5]
```

```python
# [_ for _ in arange(0, wmax, .2)]
```

```python
def bin_assigner(x, bins=bins):
    # NOTE: faster version: do a binary search on intervals
    for k, (a, b) in enumerate(bins):
        if x >= a and x < b:
            return bins[k]
    else:
        return (wmax, np.inf) # everything else
```

```python
# a little slower than the pandas version
gp_ = df.copy()
gp_['bins'] = gp_.Wind_ms.apply(bin_assigner)
gp_.head()
```

```python
gp = gp_[['Power_kw', 'bins']].groupby('bins')
power_med = gp.quantile(0.5)
power_med['q10'] = gp.quantile(0.1)
power_med['q90'] = gp.quantile(0.9)
```

```python
# pretty good replication of the power vs wind profile
power_med.plot(rot=45)
```

```python
power_med.index[:5]
```

```python
idx = bin_assigner(5)
```

```python
# need to user brackets
power_med.Power_kw.loc[ [ (0.0, 0.2) ] ]
```

```python
# it's a inverse lookup
def get_power(ws, df = power_med):
    if ws < 0:
        raise 'Not a valid ws number'
    else:
        bin_idx = bin_assigner(ws)
        return power_med.loc[[bin_idx], 'Power_kw'].values[0]
    
get_power(5.)
```

```python
# recreate the plot
wind_range = np.arange(0, wmax, .2)
power_series = pd.Series([get_power(w) for w in wind_range], index=wind_range)
power_series.plot()
```

```python
# another more advanced option -- using advanced Pandas functionality
# look-up table mean and standard deviation
bins = arange(0, wmax, .2)

# group data by bins
# advanced Pandas feature:
cuts = pd.cut(df.Wind_ms, bins)
gp = df.Power_kw.groupby(cuts) # iterator

lutm = gp.mean()
luts = gp.std()

# convert to an interval index
lutm.index = pd.IntervalIndex(lutm.index)
luts.index = pd.IntervalIndex(luts.index)
```

```python
# this is the basis of our look-up table
lutm.head(20)
```

Clean it up a bit using our knowledge of the system:

```python
# round power to nearest 5 watts == 0.05 kW
lutm = lutm.round(2)

# set final power output to 900
lutm.loc[15.8:] = 900

# make zero interval open at both ends to cover zero wind speed
lutm.loc[pd.Interval(0,.2, closed='right')] = 0 

# extend range to gusts of wind that are greater than max operating speed (25m/s) but of short duration
lutm.loc[pd.Interval(15.8, 50)] = 900
```

### Advanced solution (for further study)

```python
# use bin centres for wind speed as we will be doing linear interpolation between lookup table values
wind_ms = lutm.index.mid.values

# normalise power
power_kw = lutm.values / lutm.max()
```

```python
@vectorize
def turbine_max_power_kw(wind_speed_ms, max_power_kw=900, max_wind_speed_ms=25):
    """Linear interpolation of wind-to-power look-up table."""    
    if isnan(wind_speed_ms):
        return nan
    
    if wind_speed_ms > max_wind_speed_ms:
        return 0
    
    right_ind = searchsorted(wind_ms, wind_speed_ms)
    
    if right_ind == 0:
        return 0
    
    left_ind = right_ind - 1
    
    left_wind = wind_ms[left_ind]
    right_wind = wind_ms[right_ind]

    left_pow = power_kw[left_ind]
    right_pow = power_kw[right_ind]
    
    grad = (right_pow - left_pow) / (right_wind - left_wind)
    
    return (left_pow + grad * (wind_speed_ms - left_wind)) * max_power_kw
```

```python
# compare lookup table with data
figsize(15,5)
ws = linspace(0,30,1000)
s = step(lutm.index.right, lutm, color='r')
l = plot(ws, turbine_max_power_kw(ws, max_wind_speed_ms=25))

legend([s[0],l[0]], ['look-up table', 'turbine power function'], loc='upper left')
```

## Curtailment model

To understand how much energy is being curtailed, we need to have a model for local load and the theoretical maximum wind output that the wind generators in Orkney could produce if uncurtailed. As we are unsure of the total installed capacity of wind, we will leave these values as model parameters, to be tweaked and updated as we learn more. 

We will build a maximum, theoretical, wind-turbine power series by applying our wind-to-power model to the wind speed data. We will then normalise this (to range [0,1]) so that we only need to multiply it by our installed wind capacity parameter in order to convert it to actual wind power output (in MW). 

To model local load, we will assume that it is all residential in nature and use our aggregate load data to give us the shape of the series. After normalisation (to range [0,1]) we just have to multiply the series by our winter-peak-load in order to convert it to an actual local load value (in MW).

```python
# some suggested model parameters (you can change them)
baseload_capacity_mw = 30
wind_capacity_mw = 50
peak_load_mw = 40
interconnector_capacity_mw = 40
```

Looking at the turbine data, we seem to have about 2 years of good data (2016 and 2017) with some gaps. Looking at the load data we have one year of good data (2017). We will therefore use 2017 as our simulation year.

```python
# create simulation datetime index
ind = pd.date_range(start='2017-01-01 00:00', end='2018-01-01 00:00', freq='1min')
```

```python
# create dataframe for simulation data
sim = pd.DataFrame(index=ind)
sim.head()
```

```python
tt.Power_kw
```

```python
# add normalised actual power
tt_copy = tt.copy()
tt_copy.index = tt.index.floor('Min')
tt_copy[tt_copy.index.duplicated()] = np.NAN # tt has about 1 million rows
tt_copy.dropna(inplace=True)
```

```python
tt.shape
```

```python
# add normalised actual power
sim['Power_actual'] = tt_copy.Power_kw / 900
```

```python
sim.head()
```

```python
# add wind speed
sim['Wind_ms'] = tt_copy.Wind_ms
```

```python
# add normalised max power
sim['Power_max'] = turbine_max_power_kw(sim.Wind_ms.values, max_power_kw=1, max_wind_speed_ms=25)
```

```python
# add normalised load data
sim['Load'] = dmd.Demand_mean_kw / dmd.Demand_mean_kw.max()
```

```python
# load is sampled at 30min intervals but we need 1min interval sampling, so we will interpolate load
sim.Load = sim.Load.interpolate()
```

```python
sim[['Load', 'Power_actual', 'Power_max']].plot(alpha=.4);
sim[['Wind_ms']].plot(alpha=.8);
```

```python
# sense check
print(f'Curtailed wind capacity factor is {sim.Power_actual.mean():.2f}')
print(f'Uncurtailed wind capacity factor would be {sim.Power_max.mean():.2f}')
```
