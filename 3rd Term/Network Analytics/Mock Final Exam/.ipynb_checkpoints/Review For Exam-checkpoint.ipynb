# Creating Network

#### Creating the Network from the edge list in excel

```python
excelfile = "HW2_who_talks_to_whom.xlsx"
sent = pd.read_excel(excelfile, sheet_name="Sent", usecols = "B:CD")
received = pd.read_excel(excelfile, sheet_name="Received", usecols = "B:CD")
sent.replace({'-': np.nan}, inplace = True)
received.replace({'-': np.nan}, inplace = True) 
sent.index = np.arange(1, len(sent) + 1)
received.index = np.arange(1, len(received) + 1)
totalLen = len(sent) if len(sent) > len(received) else len(sent)
sent = sent.stack().reset_index()
sent.columns = ['from','to','talks']
received = received.stack().reset_index()
received = received[['level_1','level_0',0]]
received.columns = ['from','to','talks']
total = pd.concat([sent,received], axis = 0, join = 'outer')
total = total [['from','to','talks']]
total.columns = ['nodeA','nodeB','talks']

# Create a direct graph based on the dataframe format
dG = nx.from_pandas_edgelist(total, 'nodeA', 'nodeB', 'talks', create_using = nx.DiGraph())

# Transform node type to string, as it was shown as float
mapping = {x: str(x) for x in range(1, totalLen + 1)} 
dG = nx.relabel_nodes(dG, mapping)

undG = nx.from_pandas_edgelist(total, 'nodeA', 'nodeB', 'talks', create_using = nx.Graph())
undG = nx.relabel_nodes(undG, mapping)
```

# Degree, Centrality, Cluster

#### In-degree and out-degree for Each node

```python
d_in = dict(dG.in_degree())
d_out = dict(dG.out_degree())
sortedin = sorted(d_in.items(), key = operator.itemgetter(1), reverse = True)
sortedout = sorted(d_out.items(), key = operator.itemgetter(0), reverse = True)
l1 =list()
l2 = list()

for i in range (0,4):
     l1.append(sortedin[i][0])
     l2.append(sortedout[i][0])
    
print("The 4 people(nodes) who sent the most messages to any other person in the class are:" + " " + str(l2)) 
print("The 4 people(nodes) who received the most messages from any other person in the class are:" + " " + str(l1)) 
```

#### Betweenness centrality: 
how many times a node is in the middle of any shortest path between other nodes or put it simply, in how many paths between two other nodes (u and v for instance), this node belongs.

```python
d = dict(nx.betweenness_centrality(dG, weight = 'talks'))
sortedbet = sorted(d.items(), key = operator.itemgetter(1), reverse = True)
print("The person(node) with the highest betweeness centrality is:" + " " + str(sortedbet[0][0]))
```

#### Closeness centrality
a local metric of centrality that gives a score to each node based on how close this node is to all other nodes in the network.

```python
d = dict(nx.closeness_centrality(dG))
sortedclos = sorted(d.items(), key = operator.itemgetter(1), reverse = True)
print("The person(node) with the highest closeness centrality value is:" + " " + str(sortedclos[0][0]))
```

#### Eigenvector centrality
a global metric that measures the importance of each node which is proportional to the sum of neighbors’ importance.

```python
d = dict(nx.eigenvector_centrality_numpy(dG, weight = 'talks'))
sortedeigen = sorted(d.items(), key = operator.itemgetter(1), reverse = True)
print ("The person(node) with the highest eigenvector centrality value is:" + " " + str(sortedeigen[0][0]))
```

#### Clustering coefficient and average clustering coefficient
show how many neighbors of a node communicate each other.

```python
dcc = dict(nx.clustering(undG, weight = 'talks')) 
sortedcc = sorted(dcc.items(), key = operator.itemgetter(1), reverse = True)
print ("The person(node) with the highest clustering coefficient value is:" + " " + str(sortedcc[0][0]))

average_cc = nx.average_clustering(undG, weight = 'talks')
print("The average clustering coefficient of the graph is:" + " "+ str(average_cc))
```
