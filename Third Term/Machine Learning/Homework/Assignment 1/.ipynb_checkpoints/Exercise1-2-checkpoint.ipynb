```python
import pandas as pd
import numpy as np
import math 

from sklearn import preprocessing


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##1. load the first 800 instances of the data file
wine_data = pd.read_csv("winequality-white.csv",delimiter=";", nrows =800 ) 
```

```python
##2. construct a new binary column “good wine" which would be true if quality >=7
wine_data["good wine"] = np.where(wine_data['quality']>=7, True, False)
```

```python
wine_data
```

```python
##3.splits the data into a training set(first 400 samples), a validation set (next 200 samples), test set (last 200 samples) - no suffle needed
training_data, validation_data, test_data = wine_data[:400], wine_data[400:600], wine_data[600:]
training_data, validation_data, test_data
```

```python
## 4.normalises the data according to the Z-score transform;
columns_list = [col for col in list(training_data.columns) if col != "good wine"]
col_to_scale = training_data[columns_list]
#col_to_scale
scaler = preprocessing.StandardScaler()
training_data_scaled_array = scaler.fit_transform(col_to_scale)
training_data_scaled = pd.DataFrame(training_data_scaled_array,columns = columns_list)
training_data_scaled ## merge with good wine??
```
