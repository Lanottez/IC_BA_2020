```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Question 4.a
data = pd.read_csv('HW1_asset_prices.csv').set_index('Date')
correlation_matrix = data.corr()
correlation_matrix.head()
```

```python
# Qeuestion 4.b
stock_name = list(correlation_matrix.columns)
G = nx.Graph()
G = nx.from_numpy_matrix(correlation_matrix.values)
nx.draw_circular(G, with_labels = True)
# add title
plt.title('Circular Layout',  fontsize = 20)
plt.show()
```
