```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from datetime import datetime
import statsmodels.graphics.tsaplots as tsa
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller
from statsmodels.tools.eval_measures import rmse, aic
from statsmodels.tsa.vector_ar.irf import IRAnalysis
```

## 1. Introduction

```python
resource = pd.read_excel("Dataset/resource.xlsx")
resource.head()
```

## 2. Data

```python
plt.plot(resource.Sales, c='blue')
plt.title('Sales')
```

```python
plt.plot(resource.Adwords, c='darkgreen')
plt.title('Adwords')
```

```python
plt.plot(resource.Flyer, c='red')
plt.title('Flyers')
```

## 3. VAR Model Estimation

### 3.1 Pre-estimation Diagnostics

#### 3.1.1 Log Transformation

```python
resource_log = resource.copy()
resource_log['Sales'] = resource_log.apply(lambda row: math.log(row['Sales']+1), axis=1)
resource_log['Adwords'] = resource_log.apply(lambda row: math.log(row['Adwords']+1), axis=1)
resource_log['Flyer'] = resource_log.apply(lambda row: math.log(row['Flyer']+1), axis=1)
```

#### 3.1.2 Plot the Data to Inspect Trend

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))
tsa.plot_acf(resource['Sales'], zero = False, ax=ax[0])
tsa.plot_pacf(resource['Sales'],zero = False, ax=ax[1])
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))
tsa.plot_acf(resource['Adwords'], zero = False, ax=ax[0])
tsa.plot_pacf(resource['Adwords'],zero = False, ax=ax[1])
plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))
tsa.plot_acf(resource['Flyer'], zero = False, ax=ax[0])
tsa.plot_pacf(resource['Flyer'],zero = False, ax=ax[1])
plt.show()
```

#### 3.1.3 Unit Root Testing

```python
result = adfuller(resource_log.Sales)
print('data:  LSales')
print('Dickey-Fuller = {:.2f},Lag order = {}, p-value = {:.2f}'.format(result[0],result[2],result[1]))
print('alternative hypothesis: stationary')
```

```python
result = adfuller(resource_log.Adwords)
print('data:  LAdwords')
print('Dickey-Fuller = {:.2f},Lag order = {}, p-value = {:.2f}'.format(result[0],result[2],result[1]))
print('alternative hypothesis: stationary')
```

```python
result = adfuller(resource_log.Flyer)
print('data:  LFlyer')
print('Dickey-Fuller = {:.2f},Lag order = {}, p-value = {:.2f}'.format(result[0],result[2],result[1]))
print('alternative hypothesis: stationary')
```

```python
resource_D = resource_log.copy()
resource_D['Flyer'] = resource_D['Flyer'] - resource_D['Flyer'].shift(1)
resource_D = resource_D.dropna()
```

```python
result = adfuller(resource_log.Sales)
print('data:  LSales')
print('Dickey-Fuller = {:.2f},Lag order = {}, p-value = {:.2f}'.format(result[0],result[2],result[1]))
print('alternative hypothesis: stationary')
```

```python
result = adfuller(resource_log.Adwords)
print('data:  LAdwords')
print('Dickey-Fuller = {:.2f},Lag order = {}, p-value = {:.2f}'.format(result[0],result[2],result[1]))
print('alternative hypothesis: stationary')
```

```python
result = adfuller(resource_D.Flyer)
print('data:  LFlyer')
print('Dickey-Fuller = {:.2f},Lag order = {}, p-value = {:.2f}'.format(result[0],result[2],result[1]))
print('alternative hypothesis: stationary')
```

```python
d = {'Sales': resource_log.Sales[0:-1].tolist(), 'Adwords': resource_log.Adwords[0:-1].tolist(), 'Flyer' : resource_D.Flyer.tolist()}
data_ts_d1 = pd.DataFrame(data =d)
```

### 3.2 Construct a VAR model

```python
varp  = VAR(data_ts_d1)
results = varp.fit(1)
results.summary()
```

```python
sales_residuals = results.resid.Sales
round(sales_residuals.mean(),2)
plt.plot(sales_residuals)
```

### 3.3 IRF Analysis

```python
IRA = IRAnalysis(results)
IRA.plot(impulse='Adwords',response = 'Sales',figsize=(5,5),orth=True,repl=100)
```

```python
IRA.plot(impulse='Flyer',response = 'Sales',figsize=(5,5),orth=True,repl=100)
```

### 3.4 Optimal Allocation between Adwords and Flyer Spending

```python
cost_adwords = sum(resource.Adwords)
cost_flyer = sum(resource.Flyer)
cost_total = cost_adwords + cost_flyer

costshare_adwords = cost_adwords/cost_total
costshare_flyer = cost_flyer/cost_total
```

```python
# Ingredients for the pie-chart 
slices_actual = [costshare_adwords, costshare_flyer]
lbls_actual = ['Adwords', 'Flyer']
slices_actual[0] = round(slices_actual[0]*100)
slices_actual[1] = round(slices_actual[1]*100)

# lbls_actual = paste(lbls_actual, pct_actual)          # add data to labels
# lbls_actual = paste(lbls_actual, "%", sep="")  # add % sign to labels
labels = ['Adwords 15%','Flyer 85%']
# Get the pie-chart
plt.pie(slices_actual,labels = labels)
```
