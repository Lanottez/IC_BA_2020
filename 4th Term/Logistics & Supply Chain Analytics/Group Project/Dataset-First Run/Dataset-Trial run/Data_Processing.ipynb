```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Demand = pd.read_csv('Demand.csv')
Lost_Demand = pd.read_csv('Lost demand.csv')
```

```python
Demand = Demand.rename(columns = {'Calopeia':'Demand'})
Lost_Demand = Lost_Demand.rename(columns = {'Calopeia':'Lost_Demand'})
Demand['Lost_Demand'] = Lost_Demand.Lost_Demand
All_Data = Demand.copy()
```

```python
Finished_Goods_Inventory = pd.read_csv('Finished Goods Inventory.csv')
# Finished Goods Inventory at or destined for the warehouse in Calopeia, by current location
Finished_Goods_at_warehouse_in_Calopeia = Finished_Goods_Inventory[['day','warehouse']]
Finished_Goods_destined_for_the_warehouse_in_calopeia_by_mail = Finished_Goods_Inventory[['day','mail']]
Finished_Goods_destined_for_the_warehouse_in_calopeia_by_truck = Finished_Goods_Inventory[['day','truck']]
```

```python
Finished_Goods_at_warehouse_in_Calopeia = Finished_Goods_at_warehouse_in_Calopeia.dropna()
Finished_Goods_at_warehouse_in_Calopeia['day'] = Finished_Goods_at_warehouse_in_Calopeia.apply(lambda row:round(row['day']),axis=1)
data = Finished_Goods_at_warehouse_in_Calopeia.groupby(by = ['day'])['warehouse'].sum()
Finished_Goods_at_warehouse_in_Calopeia_processed = pd.DataFrame(data = data).reset_index()
All_Data = pd.merge(All_Data,Finished_Goods_at_warehouse_in_Calopeia_processed,how='left',on=['day'])
All_Data = All_Data.rename(columns={'warehouse':'inventory_in_warehouse'})
```

```python
Finished_Goods_destined_for_the_warehouse_in_calopeia_by_mail = Finished_Goods_destined_for_the_warehouse_in_calopeia_by_mail.dropna()
Finished_Goods_destined_for_the_warehouse_in_calopeia_by_mail['day'] = Finished_Goods_destined_for_the_warehouse_in_calopeia_by_mail.apply(lambda row:round(row['day']),axis=1)
data = Finished_Goods_destined_for_the_warehouse_in_calopeia_by_mail.groupby(by = ['day'])['mail'].sum()
Finished_Goods_destined_for_the_warehouse_in_calopeia_by_mail_processed = pd.DataFrame(data = data).reset_index()
All_Data = pd.merge(All_Data,Finished_Goods_destined_for_the_warehouse_in_calopeia_by_mail_processed,how='left',on=['day'])
```

```python
Finished_Goods_destined_for_the_warehouse_in_calopeia_by_truck = Finished_Goods_destined_for_the_warehouse_in_calopeia_by_truck.dropna()
Finished_Goods_destined_for_the_warehouse_in_calopeia_by_truck['day'] = Finished_Goods_destined_for_the_warehouse_in_calopeia_by_truck.apply(lambda row:round(row['day']),axis=1)
data = Finished_Goods_destined_for_the_warehouse_in_calopeia_by_truck.groupby(by = ['day'])['truck'].sum()
Finished_Goods_destined_for_the_warehouse_in_calopeia_by_truck_processed = pd.DataFrame(data = data).reset_index()
All_Data = pd.merge(All_Data,Finished_Goods_destined_for_the_warehouse_in_calopeia_by_truck_processed,how='left',on=['day'])
```

```python
Inventory_in_factory = pd.read_csv('WIP inventory in the Calopeia factory.csv')
Inventory_in_factory = Inventory_in_factory.rename(columns={'Calopeia':'Inventory_in_factory'})
Inventory_in_factory['day'] = Inventory_in_factory.apply(lambda row: round(row['day']),axis=1)
data = Inventory_in_factory.groupby(by='day')['Inventory_in_factory'].sum()
Inventory_in_factory_processed = pd.DataFrame(data)
All_Data = pd.merge(All_Data,Inventory_in_factory_processed,how='left',on=['day'])
```

```python
Shipment_from_warehouse = pd.read_csv('Shipments from the warehouse.csv')
Shipment_from_warehouse = Shipment_from_warehouse.rename(columns={'Calopeia':'Shipments from the warehouse'})
All_Data = pd.merge(All_Data,Shipment_from_warehouse,how='left',on=['day'])
```

```python
Cash_in_thousand = pd.read_csv('Cash balance (thousands of dollars).csv')
All_Data = pd.merge(All_Data,Cash_in_thousand,how='left',on=['day'])
```

```python
All_Data = All_Data.fillna(0)
All_Data = All_Data.set_index('day')
All_Data
```

```python
All_Data['Cash_Flow'] = All_Data.Cash.diff()
All_Data
```

```python
All_Data.to_csv("All Data.csv")
```
